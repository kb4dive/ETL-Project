```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
happiness_df = pd.read_csv("HappinessAlcoholConsumption.csv")
happiness_df = pd.DataFrame(happiness_df) 
happiness_df.head()
```

```python
happiness_df.drop(['Region', 'Hemisphere'], axis=1, inplace=True)
#happiness_df.set_index("Country", inplace=True)
happiness_df.head()
```

```python
countries_df = pd.read_csv("country_profile_variables.csv")
countries_df.head()
```

```python
countries_df = countries_df[['country', 'Population in thousands (2017)', 'Sex ratio (m per 100 f, 2017)', 
                             'Employment: Industry (% of employed)', 'Unemployment (% of labour force)']].copy()
countries_df.head()
```

```python
countries_df.rename(columns = {"country": "Country"}, inplace=True)
countries_df.rename(columns = {"Population in thousands (2017)":"Population"}, inplace=True)
countries_df.rename(columns = {"Sex ratio (m per 100 f, 2017)":"SexRatio"}, inplace=True) 
countries_df.rename(columns = {"Employment: Industry (% of employed)":"EmploymentPct"}, inplace=True)
countries_df.rename(columns = {"Unemployment (% of labour force)":"UnemploymentPct"}, inplace=True)
#countries_df.set_index("Country", inplace=True)
countries_df.head()
```

```python
suicide_df = pd.read_csv("suicide-by-country.csv")
suicide_df.head()
```

```python
suicide_df.drop(['country-year', 'generation', 'HDI for year'], axis=1, inplace=True)
suicide_df.rename(columns = {"country": "Country"}, inplace=True)
suicide_df.rename(columns = {"suicides/100k pop":"suicides_100k_pop"}, inplace=True)
suicide_df.rename(columns = {" gdp_for_year ($) ":"gdp_for_year"}, inplace=True)
suicide_df.rename(columns = {"gdp_per_capita ($)":"gdp_per_capita"}, inplace=True)
is_2015 =  suicide_df['year']==2015
suicide_df_2015 = suicide_df[is_2015]
#suicide_df_2015.set_index("Country", inplace=True)
suicide_df_2015.head()
```

```python
#Connect to database
connection_string = "postgres:postgres@localhost:5432/ETL"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#check for tables
engine.table_names()
```

```python
happiness_df.to_sql(name='happiness', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from happiness', con=engine).head()
```

```python
suicide_df_2015.to_sql(name='suicide', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from suicide', con=engine).head()
```

```python
countries_df.to_sql(name='country_profile', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from country_profile', con=engine).head()
```
